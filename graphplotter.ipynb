```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("TrainDataSets.csv")

# Set style
sns.set(style="whitegrid")

# Create a list of parameter pairs to compare
parameter_pairs = [
    ('encoder_latency_ms_A', 'encoder_latency_ms_B'),
    ('cpu_usage_percent_A', 'cpu_usage_percent_B'),
    ('packet_loss_percent_A', 'packet_loss_percent_B'),
    ('jitter_ms_A', 'jitter_ms_B'),
    ('current_fps_A', 'current_fps_B'),
    ('avg_interval_A_ms', 'avg_interval_B_ms'),
    ('adjust_bitrate_kbps_A', 'adjust_bitrate_kbps_B')
]

# Create a figure with subplots
fig, axs = plt.subplots(len(parameter_pairs), 1, figsize=(10, len(parameter_pairs)*3))

# Loop through each pair and plot
for i, (param_A, param_B) in enumerate(parameter_pairs):
    axs[i].plot(df[param_A], label=param_A, color='blue')
    axs[i].plot(df[param_B], label=param_B, color='orange')
    axs[i].set_title(f"{param_A} vs {param_B}")
    axs[i].set_ylabel("Value")
    axs[i].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("TrainDataSets.csv")

# Set rolling window size for smoothing
window_size = 20

# Parameters to compare
params = [
    ('encoder_latency_ms_A', 'encoder_latency_ms_B'),
    ('cpu_usage_percent_A', 'cpu_usage_percent_B'),
    ('packet_loss_percent_A', 'packet_loss_percent_B'),
    ('jitter_ms_A', 'jitter_ms_B'),
    ('current_fps_A', 'current_fps_B'),
    ('adjust_bitrate_kbps_A', 'adjust_bitrate_kbps_B')
]

# Plotting
plt.figure(figsize=(12, len(params)*3))
for i, (param_A, param_B) in enumerate(params, 1):
    plt.subplot(len(params), 1, i)
    plt.plot(df[param_A].rolling(window_size).mean(), label=param_A, color='blue')
    plt.plot(df[param_B].rolling(window_size).mean(), label=param_B, color='orange')
    plt.title(f"Smoothed Curve: {param_A} vs {param_B}")
    plt.xlabel("Sample Index")
    plt.ylabel("Smoothed Value")
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters to compare
kde_params = [
    ('encoder_latency_ms_A', 'encoder_latency_ms_B'),
    ('cpu_usage_percent_A', 'cpu_usage_percent_B'),
    ('packet_loss_percent_A', 'packet_loss_percent_B'),
    ('jitter_ms_A', 'jitter_ms_B'),
    ('adjust_bitrate_kbps_A', 'adjust_bitrate_kbps_B')
]

# KDE plots
plt.figure(figsize=(12, len(kde_params)*3))
for i, (param_A, param_B) in enumerate(kde_params, 1):
    plt.subplot(len(kde_params), 1, i)
    sns.kdeplot(df[param_A], label=param_A, shade=True, color="blue")
    sns.kdeplot(df[param_B], label=param_B, shade=True, color="orange")
    plt.title(f"KDE Distribution: {param_A} vs {param_B}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()

plt.tight_layout()
plt.show()
```
