```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target  # Add the target variable (median house value)
print(df)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

# Plot histogram
sns.histplot(df['MedHouseVal'], kde=True)
plt.title('Distribution of Median House Value')
plt.show()

# Q-Q plot
stats.probplot(df['MedHouseVal'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Median House Value')
plt.show()
```

```python
from scipy.stats import shapiro

# Shapiro-Wilk test
stat, p = shapiro(df['MedHouseVal'])
print(f"Shapiro-Wilk Test: Statistic={stat}, p-value={p}")

if p > 0.05:
    print("The data is normally distributed.")
else:
    print("The data is NOT normally distributed.")
```

```python
from scipy.stats import boxcox

df['MedHouseVal_boxcox'], _ = boxcox(df['MedHouseVal'] + 1)  # Add 1 to handle zeros


# Shapiro-Wilk test on transformed data
stat, p = shapiro(df['MedHouseVal_boxcox'])
print(f"Shapiro-Wilk Test: Statistic={stat}, p-value={p}")

if p > 0.05:
    print("The transformed data is normally distributed.")
else:
    print("The transformed data is NOT normally distributed.")
```

```python
# Import necessary libraries
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
from scipy.stats import boxcox, shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target  # Add the target variable (median house value)

# Apply Box-Cox transformation
# Note: Box-Cox requires strictly positive data, so we add a small constant (1) to handle zeros
#df['MedHouseVal_boxcox'], lambda_ = boxcox(df['MedHouseVal'] + 1)
from scipy.stats import yeojohnson

import numpy as np

df['MedHouseVal_boxcox'], lambda_ = yeojohnson(df['MedHouseVal'])
print(f"Optimal lambda value for Yeo-Johnson transformation: {lambda_}")


# Visualize the original and transformed data
plt.figure(figsize=(12, 6))

# Original data
plt.subplot(1, 2, 1)
sns.histplot(df['MedHouseVal'], kde=True, color='blue')
plt.title('Original Data Distribution')

# Transformed data
plt.subplot(1, 2, 2)
sns.histplot(df['MedHouseVal_boxcox'], kde=True, color='green')
plt.title('Box-Cox Transformed Data Distribution')

plt.tight_layout()
plt.show()

# Q-Q plot for the transformed data
plt.figure(figsize=(6, 6))
stats.probplot(df['MedHouseVal_boxcox'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Box-Cox Transformed Data')
plt.show()

# Shapiro-Wilk test for normality on the transformed data
stat, p = shapiro(df['MedHouseVal_boxcox'])
print(f"Shapiro-Wilk Test on Transformed Data: Statistic={stat}, p-value={p}")

if p > 0.05:
    print("The transformed data is normally distributed.")
else:
    print("The transformed data is NOT normally distributed.")
```
